```python
from sklearn.cluster import DBSCAN

from sklearn.datasets import make_moons

import matplotlib.pyplot as plt
```

```python
X,y=make_moons(n_samples=300,
               noise=0.05) #tw half circles
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
# standardscaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

X_scaled=scaler.fit_transform(X)
```

```python
dbscan=DBSCAN(eps=0.08) #domain expert ususally i am jsut playing
dbscan.fit(X_scaled)
dbscan.labels_
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=dbscan.labels_)
'''
You know see clusters but also meaning full inside the cluster thats where dbscan comes in 
meaningful insights inside clusters using Minpts and eps
'''
```
