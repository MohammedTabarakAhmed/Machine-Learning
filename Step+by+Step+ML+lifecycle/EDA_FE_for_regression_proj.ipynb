```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load dataset
df=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
df.head()
```

```python
#Data cleaning
df[df.isnull().any(axis=1)]
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
df[['Region']]=df[['Region']].astype(int)
```

```python
df
```

```python
#Removing null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]] #repeated column names
```

```python
df=df.drop(122).reset_index(drop=True) 
```

```python
df.iloc[[122]]
```

```python
df.columns #it has some spaces between them
```

```python
#fix the speaces in columns
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

### Change the required columns dtypes to int and classes as object only

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
df
```

```python
objects=[features for features in df.columns if df[features].dtypes==object]
```

```python
objects
```

```python
for i in objects:
    if i!='Classes': #except the classes column
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
#lets save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
