```python
import numpy as  np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X=6*np.random.rand(100,1)-3
y=0.5*X**2+1.5*X+2+np.random.rand(100,1)
```

```python
#visual
plt.scatter(X,y,color='y')
plt.xlabel('X dataset')
plt.ylabel('y dataset')
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#Lets implement SLR
from sklearn.linear_model import LinearRegression
regression_1=LinearRegression()
regression_1.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,regression_1.predict(X_test))
print(score)
```

```python
#lets visualize this model
plt.plot(X_train,regression_1.predict(X_train),color='r') #so we are plotting against x train vs what the predict() of x train is from the training model
plt.scatter(X_train,y_train)
plt.xlabel('X dataset')
plt.ylabel('y dataset')
```

```python
#now increase degree applying polynomial transformation
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(X_train_poly,y_train)
y_pred=regression.predict(X_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

```python
plt.scatter(X_train,regression.predict(X_train_poly))
plt.scatter(X_train,y_train)
```
