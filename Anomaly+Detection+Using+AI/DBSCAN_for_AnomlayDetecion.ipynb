```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles
X,y=make_circles(n_samples=750,factor=0.5,noise=0.7) #factor close to 1 means the circles are sae size the data point and if 0 then not same size circles(datapoints)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.10)
```

```python
dbscan.fit_predict(X)
```

```python
dbscan.labels_ #-1 outliers
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
```
