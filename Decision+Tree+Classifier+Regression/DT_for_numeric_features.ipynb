```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
#I/P features
X=pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
```

```python
#D/P features
y=iris['target']
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
#Apply Decision Tree Classififer
from sklearn.tree import DecisionTreeClassifier
treeclassifier=DecisionTreeClassifier()

treeclassifier.fit(X_train,y_train)
```

```python
#Visualize the Decision Tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
y_pred=treeclassifier.predict(X_test)
```

```python
#metrics
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred)) #if you remove max+depth the accuracy increases
```

```python
import pickle

with open("tree_model.pkl", "wb") as f:
    pickle.dump(treeclassifier, f)
```

```python
data_bundle = {
    "X_train": X_train,
    "X_test": X_test,
    "y_train": y_train,
    "y_test": y_test
}

# save to pickle file
with open("train_test_bundle.pkl", "wb") as f:
    pickle.dump(data_bundle, f)
```
