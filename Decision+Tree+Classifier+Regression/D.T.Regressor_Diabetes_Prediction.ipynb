```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes
df=load_diabetes()
```

```python
print(df['DESCR'])
```

```python
#create dataframe
df_diabetes=pd.DataFrame(df.data,columns=)
```

```python
df_diabetes=pd.DataFrame(df.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
df_diabetes.head()
```

```python
#I/P and D/P features
X=df_diabetes
y=df['target']
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
#CHECK CORRELATION  
X_train.corr()
```

```python
#Heatmap corr()
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
#DecisionTreeRegressor(fit)

from sklearn.tree import DecisionTreeRegressor
treeregressor=DecisionTreeRegressor()

treeregressor.fit(X_train,y_train)
```

```python
#Hyperparamter tuning
param_grid = {
    "criterion": ["squared_error", "absolute_error", "friedman_mse"],
    'splitter':['best','random'],
    "max_depth": [None, 3, 5, 10],
    "min_samples_split": [2, 5, 10]
}
```

```python
#GridSearchCV
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(treeregressor,cv=5,param_grid=param_grid,scoring='neg_mean_squared_error')

grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
#metrics
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

print("R2:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:",np.sqrt((y_test, y_pred,)))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MAPE:", mean_absolute_percentage_error(y_test, y_pred))
```

```python
#Model with grid search cv

selected_model = DecisionTreeRegressor(
    criterion="squared_error",
    max_depth=5,
    min_samples_split=5,
    splitter="random",
    random_state=42
)
selected_model.fit(X_train, y_train)
```

```python
#visualize the DTR
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selected_model,filled=True)
```
